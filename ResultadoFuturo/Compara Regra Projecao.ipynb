```python
urlGeral = "http://vps-40d69db1.vps.ovh.ca:22001/api/CotacaoIntradayAcaoResultadoValors/obtemResultadoRegraData?codigoRegra=t15s15v&diaNumInicial=20211201&diaNumFinal=20220131"
```

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
r = requests.get(urlGeral)
j = r.json()
dfStop15 = pd.DataFrame.from_dict(j)
```

```python
urlstop10 = "http://vps-40d69db1.vps.ovh.ca:22001/api/CotacaoIntradayAcaoResultadoValors/obtemResultadoRegraData?codigoRegra=t15s10v&diaNumInicial=20211201&diaNumFinal=20220131"
```

```python
r = requests.get(urlstop10)
j = r.json()
dfStop10 = pd.DataFrame.from_dict(j)
```

```python
dfMerge = pd.merge(dfStop10,dfStop15,how='inner',
                   left_on=['ano','mes','ticker'], 
                   right_on=['ano','mes','ticker'], suffixes=('_10', '_15'))
```

```python
dfSort = dfMerge.sort_values(by='lucro_10' ,ascending=False)
dfSort.head(10)
```

```python
dfGrupo = dfSort.groupby(['ano','mes']).sum()
```
