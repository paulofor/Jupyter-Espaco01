```python
url = "http://vps-40d69db1.vps.ovh.ca:21046/api/CotacaoIntradayAcaoResultadoValors/obtemListaTickerRegra?ticker=ABEV3&idRegra=1"
```

```python
import requests
import pandas as pd

r = requests.get(url)

j = r.json()

dfABEV3 = pd.DataFrame.from_dict(j)
```

```python
dfABEV3.head()
```

```python
import matplotlib.pyplot as plt
```

```python
dfABEV3.plot(x="dataHora", y=["valorEntrada","resultado", "dias"], 
             figsize=(18,5), legend=None)
plt.show()
```
