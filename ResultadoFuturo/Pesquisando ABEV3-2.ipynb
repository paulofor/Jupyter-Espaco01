```python
url = "http://vps-40d69db1.vps.ovh.ca:21046/api/CotacaoIntradayAcaoResultadoValors/obtemListaTickerRegra?ticker=ABEV3&idRegra=1"
```

```python
import matplotlib.pyplot as plt
```

```python
import requests
import pandas as pd

r = requests.get(url)

j = r.json()

dfABEV3 = pd.DataFrame.from_dict(j)
```

```python
dfABEV3.tail(5)
```

Usar apenas dados de 3 meses ( setembro, outubro e novembro)

```python
dfABEV3_3Meses = dfABEV3.drop(dfABEV3[dfABEV3.diaNum < 20210901].index)
```

```python
dfABEV3_3Meses.head(10)
```

Plotagem do Período

```python
dfABEV3_3Meses.plot(x="dataHora", y=["valorEntrada","resultado"], 
             figsize=(18,5), legend=None)
plt.show()
```
