# Obtendo Dados

```python
url = "http://vps-40d69db1.vps.ovh.ca:21046/api/CotacaoIntradayAcaoResultadoValors/obtemListaTickerRegra?ticker=ABEV3&idRegra=1"
```

# Imports

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

# Execuções

```python
r = requests.get(url)
j = r.json()
df = pd.DataFrame.from_dict(j)

now = datetime.now()
dt_string = now.strftime("%d/%m/%Y %H:%M:%S")

print(dt_string)
df.plot(x="dataHora", y=["valorEntrada","resultado", "dias"], 
             figsize=(18,5), legend=None)
plt.show()
```
