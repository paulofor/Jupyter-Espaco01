```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
r = requests.get(url)
j = r.json()
df = pd.DataFrame.from_dict(j)
```

## Média de Dias

```python
dfDias = df.groupby(['ticker']).mean(['mediaDias'])
dfDias.head()
```
