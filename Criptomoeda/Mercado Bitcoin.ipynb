```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
url = "http://vps-40d69db1.vps.ovh.ca:22001/api/CotacaoIntradayMercadoBitcoins/fechamentosPorDiaNum?diaNum=20220217"
```

```python
r = requests.get(url)
j = r.json()
df = pd.DataFrame.from_dict(j)
```

```python
df2 = df.set_index('ticker')
```

```python
df2.head()
```

```python
url1mes = "http://vps-40d69db1.vps.ovh.ca:22001/api/CotacaoIntradayMercadoBitcoins/fechamentosPorDiaNum?diaNum=20220117"
```

```python
r = requests.get(url1mes)
j = r.json()
df1mes = pd.DataFrame.from_dict(j)
```

```python
df1mes = df1mes.set_index('ticker')
```

```python
df1mes.head()
```

```python
dfmerge = pd.merge(df1mes,df2,how='inner' , left_index=True, right_index=True, suffixes=('_1', '_0'))
```

```python
dfmerge.head()
```

```python
dfmerge['lucro'] = dfmerge['valor_0'] - dfmerge['valor_1']
```

```python
dfmerge['percentual'] = (dfmerge['lucro'] / dfmerge['valor_1'] * 100)
```

```python
dfmerge.sort_values(by='percentual', ascending=False)
```
