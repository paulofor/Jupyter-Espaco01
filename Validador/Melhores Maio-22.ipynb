```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
r = requests.get(url)
j = r.json()
df = pd.DataFrame.from_dict(j)
```

```python
df.head()
```

```python
df['codigoRegraProjecao'].unique()
```

```python
def lucro(linha):
    linha['lucro'] = (linha['target']*linha['totalLucro']) - (linha['stop'] * linha['totalPrejuizo'])
    return linha
```

```python
df = df.apply(lucro, axis='columns')
```

```python
df.head()
```

```python
dfLucro = df.filter(items=['ticker','lucro'])
```

```python
dfLucro.head()
```

```python
dfTickerLucro = dfLucro.groupby(['ticker']).sum()
```

```python
dfTickerLucro.head()
```

```python
dfRank = dfTickerLucro.sort_values(by='lucro', ascending=False)
dfRank.head(10)
```
