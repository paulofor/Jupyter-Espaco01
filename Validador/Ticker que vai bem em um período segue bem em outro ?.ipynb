### Neg-100 (88 tickers)

```python
url = "http://vps-40d69db1.vps.ovh.ca:23002/api/CotacaoIntradayAcaoResultadoValors/obtemDatasPorDataGrupoLimite?diaInicioNum=20220301&diaFinalNum=20220430&grupoAcaoId=12&limiteDiaRegra=260"
```

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
r = requests.get(url)
j = r.json()
df = pd.DataFrame.from_dict(j)
```

```python
df.head()
```

```python
dfMasc = df['codigoRegraProjecao'].isin(['t10s17vy','t12s15vy','t15s12vy','t15s15vy'])
```

```python
df2 =  df[dfMasc]
```

```python
df2.head()
```

```python
def lucro(linha):
    linha['lucro'] = (linha['target']*linha['totalLucro']) - (linha['stop'] * linha['totalPrejuizo'])
    return linha
```

```python
df = df2.apply(lucro, axis='columns')
```

```python
df.head()
```

```python
dfLucro = df.filter(items=['ticker','lucro'])
```

```python
dfLucro.head()
```

```python
dfTickerLucro = dfLucro.groupby(['ticker']).sum()
```

```python
dfTickerLucro.head()
```

## Melhores para Venda

```python
dfRank = dfTickerLucro.sort_values(by='lucro', ascending=False)
dfRank.head(20)
```

## Melhores para Compra

```python
dfRank2 = dfTickerLucro.sort_values(by='lucro', ascending=True)
dfRank2.head(30)
```
