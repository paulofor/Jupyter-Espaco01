```python
url = "http://vps-40d69db1.vps.ovh.ca:23002/api/CotacaoIntradayAcaoResultadoValors/obtemDatasPorDataGrupoLimite?diaInicioNum=20220201&diaFinalNum=20220430&grupoAcaoId=12&limiteDiaRegra=260"
```

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
r = requests.get(url)
j = r.json()
df = pd.DataFrame.from_dict(j)
```

## Melhores para Venda - Por Prejuizo

```python
dfTot = df.groupby('ticker').sum(['totalLucro', 'totalPrejuizo'])
```

```python
dfRank1 = dfTot.sort_values(by='totalPrejuizo', ascending=True)
dfRank1.head(15)
```
