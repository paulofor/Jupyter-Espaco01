```python
import numpy as np
import pandas as pd
import requests
```

```python
def min_max(row):
    data = row[['positivos','negativos','zero']]
    return pd.Series({'min' : np.min(data) , 'max' : np.max(data)})
```

```python
def percentuais(linha):
    linha['lucro'] = 100 * (linha['positivos'] / linha['ticks'])
    linha['prejuizo'] = 100 * (linha['negativos'] / linha['ticks'])
    linha['nulos'] = 100 * (linha['zero'] / linha['ticks'])
    return linha
```

```python
urlOut = "http://vps-40d69db1.vps.ovh.ca:22001/api/RegraProjecaos/tickerPerformance?dataNumInicio=20211001&dataNumFinal=20211031&grupoAcao=Neg-100&codigoRegra=1"
```

```python
r = requests.get(urlOut)
j = r.json()
dfOut = pd.DataFrame.from_dict(j)
dfOut.set_index('ticker', inplace=True)
```
