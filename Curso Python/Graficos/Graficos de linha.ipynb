```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dados_linear = np.array([1,2,3,4,5,6,7,8])
dados_quadrado = dados_linear**2

plt.figure()
plt.plot(dados_linear, '-o' , dados_quadrado, '-o')
plt.plot([22,44,55], '--r')

plt.xlabel('Some data')
plt.ylabel('Some other data')
plt.title('titulo')
plt.legend(['Baseline', 'Competition', 'us'])

plt.gca().fill_between(range(len(dados_linear)),
                      dados_linear, dados_quadrado, 
                      facecolor='blue',
                      alpha=0.15)
```
