```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

```python
plt.figure()
data = np.random.rand(10)
plt.plot(data)

def onclick(event):
    plt.cla()
    plt.plot(data)
    plt.gca().set_title('Event at pixel {}, {}, and data {},{}'.format(event.X,event.Y,'\n',event.xdata,event.ydata))
    
    
plt.gcf().canvas.mpl_connect('button_press_event',onclick)    
```

```python
from random import shuffle

origins = ['China' , 'Brazil' , 'India', 'USA' , 'Canada' , 'UK' , 'Germany' , 'Iraq' , 'Chile' , 'Mexico']
shuffle(origins)

df = pd.DataFrame({'height' : np.random.rand(10) , 'weight' : np.random.rand(10) , 
                  'origins' : origins})
df
```

```python
plt.figure()
plt.scatter(df['height'], df['weight'], picker=5)
plt.gca().set_ylabel('Weight')
plt.gca().set_xlabel('Height')
```

```python
def onpick(event):
    origin = df.iloc[event.id[0]]['origin']
    plt.gca().set_title('Selected item came from {}'.format(origin))
plt.gcf().canvas.mpl_connect('pick_event', onpick)
```
