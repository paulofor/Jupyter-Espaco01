```python
import matplotlib.pyplot as plt

import numpy as np
```

```python
plt.figure()
plt.subplot(1,2,1)

dados_lineares = np.array([1,2,3,4,5,6,7,8])
plt.plot(dados_lineares, '-o')
```

```python
dados_exponenciais = dados_lineares**2

plt.subplot(1,2,2)
plt.plot(dados_exponenciais, '-o')
```

```python
plt.subplot(1,2,1)
plt.plot(dados_exponenciais, '-x')
```

```python
plt.figure()
ax1 = plt.subplot(1,2,1)
plt.plot(dados_lineares, '-o')
ax2 = plt.subplot(1,2,2, sharey=ax1)
plt.plot(dados_exponenciais, '-x')
```
